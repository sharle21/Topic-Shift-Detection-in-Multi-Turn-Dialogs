```python
import pandas as pd

# Read the file
with open(r"E:\Project_602\ijcnlp_dailydialog\train\dialogues_train.txt", 'r', encoding='utf-8') as file:
    lines = file.read().strip()

# Split the text into conversations based on newlines
conversations = lines.split('\n')

# Initialize the DataFrame structure
data = []

# Process each conversation
for conv_index, conversation in enumerate(conversations):
    messages = conversation.split('__eou__')
    messages = [msg.strip() for msg in messages if msg.strip()]  # Remove empty and whitespace lines
    
    # Process each message
    for count, message in enumerate(messages):
        agent = f'agent_{1 if count % 2 == 0 else 2}'  # Alternate agents
        data.append([count, agent, message])

# Create DataFrame
df = pd.DataFrame(data, columns=['Count', 'Agents', 'Messages'])

# Reset count for new chats (grouped for display)
df['Chat_ID'] = df['Count'].eq(0).cumsum()
df = df[['Chat_ID', 'Count', 'Agents', 'Messages']]
```
